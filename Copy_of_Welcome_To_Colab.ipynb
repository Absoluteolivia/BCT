```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Prepare the historical data (2019-2025)
# Data represents summer average temperatures in Fahrenheit
data = {
    'Year': [2019, 2020, 2021, 2022, 2023, 2024, 2025],
    'NYC_Summer_Avg_F': [75.6, 76.9, 76.0, 76.8, 74.7, 76.6, 75.5],
    'Global_Anomaly_C': [0.98, 1.02, 0.84, 0.89, 1.17, 1.18, 1.19]
}

df = pd.DataFrame(data)

# 2. Save to CSV for your assignment submission
df.to_csv('summer_temperature_assignment.csv', index=False)
print("CSV file 'summer_temperature_assignment.csv' has been created.")

# 3. Perform Linear Regression for 2026 Prediction
X = df[['Year']]
y = df['NYC_Summer_Avg_F']

model = LinearRegression()
model.fit(X, y)

# Predict for summer 2026
year_2026 = np.array([[2026]])
prediction_2026 = model.predict(year_2026)[0]

# 4. Visualization
plt.figure(figsize=(12, 6))

# Plot historical data
plt.plot(df['Year'], df['NYC_Summer_Avg_F'], marker='o', color='royalblue', label='Historical Avg (NYC)')

# Plot prediction point
plt.scatter(2026, prediction_2026, color='red', s=100, label=f'Predicted 2026: {prediction_2026:.2f}°F')

# Trend line calculation
trend_line = model.predict(np.append(df['Year'].values, 2026).reshape(-1, 1))
plt.plot(np.append(df['Year'].values, 2026), trend_line, color='orange', linestyle='--', alpha=0.6, label='Trend Line')

# Formatting the chart
plt.title('Summer Temperature Trend Analysis & 2026 Prediction', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature (°F)', fontsize=12)
plt.xticks(np.arange(2019, 2027, 1))
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()

# Show findings
plt.show()

print(f"--- Assignment Insights ---")
print(f"Historical Mean: {df['NYC_Summer_Avg_F'].mean():.2f}°F")
print(f"Probable temperature for Summer 2026: {prediction_2026:.2f}°F")
```

```
import seaborn as sns
import matplotlib.pyplot as plt
#Load Titanic dataset
df = sns.load_dataset("titanic")

 #FARE VS AGE, COLORED BY SURVIVAL
sns.scatterplot(x = "age", y = "fare", hue="survived", data=df)
plt.title("FARE VS AGE(COLORED BY SURVIVAL)")
plt.xlabel("AGE")
plt.ylabel("FARE")
plt.show()

#CORRELATION MATRIX
corr = df[['age','fare','survived']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("CORRELATION MATRIX")
plt.show()

#CHECK MISSING VALUES
print(df.isnull().sum())

#fill missing age values with mean
df['age'] = df['age'].fillna(df['age'].mean())
print(df.isnull().sum())

#FILL MISSING EMBARK_TOWN WITH 'UNKNOWN'
df['embark_town'] = df['embark_town'].fillna('Unknown')
print(df["embark_town"])

#DROP ROWS WITH MISSING DECK INFO
df = df.dropna(subset=['deck'])
print(df)
#SUrvival rate by gender
survival_by_gender = df.groupby('sex')['survived'].mean()
print(survival_by_gender)
#SUrvival rate by class
survival_by_class = df.groupby('sex')['survived'].mean()
print(survival_by_class)

#SUrvival rate by gender and class
survival_by_gender_class = df.groupby(['sex','pclass'])['survived'].mean()
print(survival_by_gender_class)

#bar plot:survival rate by gender
sns.barplot(x="sex", y="survived", data=df, ci= None) #(only df there is a confidence interval,to remove it ci=None}
plt.title("Survival Rate by Gender")
plt.show()

#HEATMAP: SURVIVAL BY GENDER & CLASS
pivot = df.pivot_table(index='sex', columns='pclass', values='survived', aggfunc='mean')

sns.heatmap(pivot, annot=True, cmap="Blues")
plt.title("Survival Rate by Gender & Class")
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
#Load Titanic dataset
df = sns.load_dataset("tips")
#preview the data
df.head()
#check missing valus
print(df.isnull().sum())
print(df.dtypes)
#convert 'sex' and 'smoker' to category type
df['sex'] = df['sex'].astype('category')
df['smoker'] = df['smoker'].astype('category')
#creat a new column: tip percentage
df['tip_percentage'] = df['tip'] / df['total_bill']*100
print(df.head())
#Round to 2 decimal places
df['tip_percentage'] = df['tip_percentage'].round(2)
print(df.head())
#preview updated data frame
df[['total_bill','tip','tip_percentage']].head()
#Average tip percentage by gender
avg_tip_percentage_by_gender = df.groupby('sex',observed = False)['tip_percentage'].mean()
print(avg_tip_percentage_by_gender)
#Average tip percentage by day and time
avg_tip_percentage_by_day_time = df.groupby(['day','time'], observed =False)['tip_percentage'].mean().unstack()
print(avg_tip_percentage_by_day_time)

import matplotlib.pyplot as plt

#bar plot: tip % by gender
sns.barplot(x="sex", y="tip_percentage", data=df)
plt.title("Avaerage Tip Percentage by Gender")
plt.show()
#bar plot: tip % by smoker status
sns.barplot(x="smoker", y="tip_percentage", data=df)
plt.title("Avaerage Tip Percentage by Gender")
plt.show()
#box plot: tip % by day
sns.boxplot(x="day", y="tip_percentage", data=df)
plt.title("Tip Percentage by Day")
plt.show()
```
